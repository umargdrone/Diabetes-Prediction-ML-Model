```python
delta, alpha_h, lambda_hxi, psi_hxi, mu_hxi, alpha_hxi, r_x, gamma_hxi, alpha_v, lambda_v, psi_v, mu_v, gamma_v, psi_h, lambda_h, nu_h, mu_h, omega, gamma_h, r, S_h, E_h, I_h, R_h, S_v, E_v, I_v = symbols('delta, alpha_h, lambda_hxi, psi_hxi, mu_hxi, alpha_hxi, r_x, gamma_hxi, alpha_v, lambda_v, psi_v, mu_v, gamma_v, psi_h, lambda_h, nu_h, mu_h, omega, gamma_h, r, S_h, E_h, I_h, R_h, S_v, E_v, I_v')
```

```python
# dShdt = psi_h -(delta*lambda_h*S_h*I_v)/(1+nu_h*I_v) - mu_h*S_h+omega*R_h
# dEhdt = (delta*lambda_h*S_h*I_v)/(1+nu_h*I_v) - (alpha_h+mu_h*E_h)
# dIhdt = alpha_h*E_h -(r+gamma_h+mu_h)*(I_h)
# dRhdt = r*I_h-(mu_h+omega)*(R_h)
# dSvdt = psi_v-(delta*lambda_v*S_v*I_h)/(1+nu_h*I_h)-mu_v*S_v
# dEvdt = (delta*lambda_h*S_h*I_v)/(1+nu_h*I_v)-(alpha_v+mu_v)*E_v
# dIvdt = alpha_v*E_v-(mu_v+gamma_v)*I_v

lambda_v = 0.758
lambda_h =0.8354
alpha_h = 0.5
alpha_v = 0.6
psi_v = 0.4
psi_h= 0.459
gamma_v = 0.085
gamma_h = 0.09
nu_h = 0.1
omega = 0.5
r=0.9
delta = 0.08
mu_v = 17.3809
mu_h = 0.01567
r = 0.0667
c_b = 0.098
sigma_b = 0.085
mu_b = mu_v
kappa_b = 0.09
```

```python
def mymodel(y, t, delta, alpha_h, lambda_v, alpha_v, psi_v, mu_v, gamma_v, psi_h, lambda_h, nu_h, mu_h, omega, gamma_h, r):
    S_h, E_h, I_h, R_h, S_v, E_v, I_v = y

    dShdt = psi_h -(delta*lambda_h*S_h*I_v)/(1+nu_h*I_v) - mu_h*S_h+omega*R_h
    dEhdt = (delta*lambda_h*S_h*I_v)/(1+nu_h*I_v) - (alpha_h+mu_h*E_h)
    dIhdt = alpha_h*E_h -(r+gamma_h+mu_h)*(I_h)
    dRhdt = r*I_h-(mu_h+omega)*(R_h)
    dSvdt = psi_v-(delta*lambda_v*S_v*I_h)/(1+nu_h*I_h)-mu_v*S_v
    dEvdt = (delta*lambda_h*S_h*I_v)/(1+nu_h*I_v)-(alpha_v+mu_v)*E_v
    dIvdt = alpha_v*E_v-(mu_v+gamma_v)*I_v
    return [dShdt, dEhdt, dIhdt, dRhdt, dSvdt, dEvdt, dIvdt]

# initial_conditions = [9000, 800, 100, 50, 800, 20, 50]
# initial_conditions = [9000, 800, 1, 0, 3, 2, 1]
initial_conditions = [8000, 50, 20, 0, 8000, 50, 20]

time_range = np.linspace(0, 150)


sol2 = odeint(mymodel, initial_conditions, time_range, args=(delta, alpha_h, lambda_v, alpha_v, psi_v, mu_v, gamma_v, psi_h, lambda_h, nu_h, mu_h, omega, gamma_h, r))

```

```python
plt.plot(time_range, sol2[:,0], label='Susceptible humans')
plt.plot(time_range, sol2[:,1], label='Exposed humans')
plt.plot(time_range, sol2[:,2], label='Infected humans')
plt.plot(time_range, sol2[:,3], label='Recovered humans')
plt.plot(time_range, sol2[:,4], label='Susceptible black-flies')
plt.plot(time_range, sol2[:,5], label='Exposed black-flies')
plt.plot(time_range, sol2[:,6], label='Infected black-flies')

# plt.plot(time_range, sol1[:,0], label='S')


plt.xlabel("Days")
# plt.ylabel("Susceptible humans to both dieseases")

plt.legend(['Susceptible humans', 'Exposed humans', 'Infected humans', 'Recovered humans', 'Susceptible black-flies', 'Exposed black-flies', 'Infected black-flies'], loc=(1,0.59))
# plt.legend(loc='upper right')
```
